#### 1. Import and combine pollution raster and school locations

```python
raster_PM25 = rasterio.open("data/AnnAvg1_8_300mSurfaces/aa3_pm300m//")

# Or download from NYC Open Data: https://data.cityofnewyork.us/Environment/NYCCAS-Air-Pollution-Rasters/q68s-8qxv
# Check the link for the data dictionary
```

```python
show(raster_PM25,title = 'NYCCAS raster showing PM2.5 concentrations (2012)');
```

```python
# Import the shapefile with school locations and air pollution / socioeconomic data

df = gp.GeoDataFrame.from_file('air/schools_all_info.shp')
df1 = df[['BORO','SCHOOLNAME','SCH_TYPE','ADDRESS','PRINCIPAL','PRIN_PH','GRADES',
          'pm_1','pm_2','pm_3','pm_4','pm_5','pm_6','pm_7','Name','Poverty','Unemployme','Self_rep_h',
         'Asthma_5to','Life_expec','geometry']]

# Cleaning: we focus on four boroughs leaving Staten Island out for now, and drop suspiciously high or low values.

df1 = df1[df.BORO != 'R']
df1=df1[(df.pm_7 > 6) & (df.pm_6 > 6) & (df.pm_5 > 6) & (df.pm_4 > 6) & (df.pm_3 > 6) & (df.pm_2 > 6) & (df.pm_1 > 6)]

print(df1.shape)
df1.head()
```

```python
# Plot the Geopandas figure
df1.plot()
```

```python
# Which school types are included?

print(df1.SCH_TYPE.value_counts())
df1.SCH_TYPE.value_counts().plot(kind="barh",title="Number of schools by type").invert_yaxis();
```

# How does air quality differ by Borough? And many schools have hazardous air quality? (PM2.5 > 10)

```python
# Histogram of 2015 PM2.5 values for elementary schools by borough

boros = ['M','K','Q','X']
boro_names = ['Manhattan','Brooklyn','Bronx','Queens']
cols = ['b','g','purple','y']

for i in range(len(boros)):

    f,ax = plt.subplots(1,figsize=[6,4])
    df[df.SCH_TYPE=='Elementary'][df.BORO==boros[i]].pm_7.hist(ax=ax,color=cols[i])
    plt.title(boro_names[i]+" elementary schools: PM2.5 levels (2015)");
    plt.xlabel("PM2.5 level (micrograms / square meter)")
    plt.ylabel("Frequency")
    plt.xlim(7,12)
    plt.axvline(x=10,linestyle='dotted',color='r')
```

```python
# Histogram of 2015 PM2.5 values for elementary schools by borough

boros = ['M','K','Q','X']
boro_names = ['Manhattan','Brooklyn','Bronx','Queens']
cols = ['b','g','purple','y']

for i in range(len(boros)):

    f,ax = plt.subplots(1,figsize=[6,4])
    df[df.SCH_TYPE=='High school'][df.BORO==boros[i]].pm_7.hist(ax=ax,color=cols[i])
    plt.title(boro_names[i]+" high schools: PM2.5 levels (2015)");
    plt.xlabel("PM2.5 level (micrograms / square meter)")
    plt.ylabel("Frequency")
    plt.xlim(7,13)
    plt.axvline(x=10,linestyle='dotted',color='r')
```

# Did PM2.5 levels improve in the past 7 years?

```python
boros = ['M','K','Q','X']
boro_names = ['Manhattan','Brooklyn','Bronx','Queens']

years = ['pm_1','pm_2','pm_3','pm_4','pm_5','pm_6','pm_7']
year_names = ['2009','2010','2011','2012','2013','2014','2015']

M_mean=[]
K_mean=[]
Q_mean=[]
X_mean=[]
S_mean=[]

# M
for i in range(len(years)):
    M_mean.append(df1[df1.BORO=='M'][years[i]].mean())

# K
for i in range(len(years)):
    K_mean.append(df1[df1.BORO=='K'][years[i]].mean())

# Q
for i in range(len(years)):
    Q_mean.append(df1[df1.BORO=='Q'][years[i]].mean())

# X
for i in range(len(years)):
    X_mean.append(df1[df1.BORO=='X'][years[i]].mean())

dict1 = {"Manhattan":M_mean,"Brooklyn":K_mean,'Queens':Q_mean,'Bronx':X_mean}
boro_means = pd.DataFrame.from_dict(dict1)
boro_means.index=year_names
```

#### Table of annual mean PM2.5 levels by borough

```python
boro_means
```

#### Plot showing that yes, PM2.5 levels have fallen

```python
fig,ax=plt.subplots(figsize=[12,7])
for i in boro_means.columns:
    plt.plot(boro_means[i],label=i)
ax.set_title("Mean schoolyard PM2.5 by borough")
ax.set_xticklabels(['2009','2010','2011','2012','2013','2014','2015','2016'])
plt.legend()
plt.axhline(y=10,linestyle='dotted',color='k')
```

# Are PM2.5 levels correlated with poverty?
* answer: no sign of it. Expensive schools in Manhattan are near a lot of traffic.
* though Bronx has some linear relationship
* could use a little more work and less important of an output.

```python
boros = ['M','K','Q','X']
boro_names = ['Manhattan','Brooklyn','Queens','Bronx']
cols = ['blue','g','purple','y']

for i in range(len(boros)):
    sns.pairplot(x_vars='Poverty',y_vars='pm_7',data=df1[df1.BORO==boros[i]],size=4)
    plt.title(boro_names[i]+' schools: PM2.5 vs poverty levels')
    plt.xlabel('Percent below poverty line')
    plt.xlim(5,45)
    plt.ylim(7,13)
    plt.ylabel('PM2.5 concentration (ug/m3)');
```
